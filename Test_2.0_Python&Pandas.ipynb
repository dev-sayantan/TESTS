```python
def fibonacci(n):
    n1, n2 = 0, 1
    count = 0
    if n < 0:
        print("Enter a positive number.")
    elif n == 1:
        print(f"Fibonacci Sequence upto {n} = ",n1)    
    else:
        print("Fibonacci sequence")
        while count < n:
            print(n1)
            next_num = n1+n2
            n1, n2 = n2, next_num
            count+=1
fibonacci(10)
```

```python
def is_prime(num):
    flag = True
    if num < 2:
        flag = False
    if num >= 2:
        for i in range(2,num):
            if num % i == 0:
                flag = False
                break
    return flag

is_prime(7)
        
```
